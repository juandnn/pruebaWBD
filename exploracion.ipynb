### En este notebook se hace la exploración de la base datos de panorama TV infantil

```python
import pandas as pd

# Cargar el archivo excel

df = pd.read_excel('OM - Base de datos panorama TV Infantil 2018 - 2019.xlsx')
df.head()
```

```python
# ver el tamaño de dataset (filas, columnas)
df.shape
```

```python
# Por cada columna, ver el tipo de dato y la cantidad de valores nulos
df.isnull().sum()
```

```python
# por cada columna ver los valores únicos y su cantidad
for column in df.columns:
    print(f"Columna: {column}")
    print(df[column].value_counts())
    print("\n")
    
```

```python
# cargar el archivo excel como csv a la carpeta /src/data 

df.to_csv('src/data/panorama_tv_infantil.csv', index=False)
```

```python
# Cargar diccionario con las descripciones

import json


column_descriptions = {
    "Canal": "Nombre del canal de televisión",
    "Franja Horaria": "Rango de horas de transmisión",
    "Año": "Año de la transmisión",
    "Fecha": "Fecha específica de la transmisión",
    "Día de la semana": "Día de la semana de la transmisión",
    "Mes": "Mes de la transmisión",
    "Share Niños 4-8 años": "Porcentaje de participación entre televidentes niños de 4-8 años",
    "Rating en Miles Niños 4-8 años": "Número en miles de niños de 4-8 años conectados al canal",
    "Share Amas de casa 18+": "Porcentaje de participación entre amas de casa de 18+ años",
    "Rating en Miles Amas de casa 18+": "Número en miles de amas de casa de 18+ años conectadas al canal",
    "Share Personas 4+": "Porcentaje de participación entre personas de 4+ años",
    "Rating en Miles Personas 4+": "Número en miles de personas de 4+ años conectadas al canal"
}

with open('src/data/diccionario_completo.json', 'w', encoding='utf-8') as f:
    json.dump(column_descriptions, f, indent=2, ensure_ascii=False)
```

# Agrupaciones por mes

```python
# crear un df nuevo con las columnas agrupadas por promedio por mes para cada canal quitando las columnas de fecha, día de la semana y franja horaria

df_grouped = df.groupby(['Canal', 'Año', 'Mes']).agg({
    'Share Niños 4-8 años': 'mean',
    'Rating en Miles Niños 4-8 años': 'sum',
    'Share Amas de casa 18+': 'mean',
    'Rating en Miles Amas de casa 18+': 'sum',
    'Share Personas 4+': 'mean',
    'Rating en Miles Personas 4+': 'sum'
}).reset_index()


df_grouped.head()
```

```python
# muestra df_grouped para discovery kids

df_grouped[df_grouped['Canal'] == 'Discovery Kids']
```

```python
# subir csv a la carpeta /src/data

df_grouped.to_csv('src/data/panorama_tv_infantil_agrupado_mes.csv', index=False)
```

```python
# Agrupación audiencia máxima por mes para Discovery Kids

df_discovery_kids = df[df['Canal'] == 'Discovery Kids'].copy()
df_discovery_kids['Total Niños 4-8 años'] = df_discovery_kids['Rating en Miles Niños 4-8 años'] / (df_discovery_kids['Share Niños 4-8 años'] / 100)
df_discovery_kids['Total Amas de casa 18+'] = df_discovery_kids['Rating en Miles Amas de casa 18+'] / (df_discovery_kids['Share Amas de casa 18+'] / 100)
df_discovery_kids['Total Personas 4+'] = df_discovery_kids['Rating en Miles Personas 4+'] / (df_discovery_kids['Share Personas 4+'] / 100)

df_totalesDiscovery = df_discovery_kids.groupby(['Año', 'Mes']).agg({
    'Total Niños 4-8 años': 'sum',
    'Total Amas de casa 18+': 'sum',
    'Total Personas 4+': 'sum'
}).reset_index().round(2)

df_totalesDiscovery.head()
```

```python
# guardar totales en csv en la carpeta /src/data

df_totalesDiscovery.to_csv('src/data/totales_discovery_kids.csv', index=False)
```
